```python
import matplotlib.pyplot as plt
import numpy
import control
```

```python
Ts = 0.1
a = 0.5
b = 0.7
G = control.tf([1, 1], numpy.convolve([1, -a],[1, -b]), Ts)
G
```

```python
t = numpy.arange(0, 2.5+Ts, Ts)
sampledt, impulse_response = control.impulse_response(G,t)
```

```python
plt.stem(sampledt, impulse_response, use_line_collection=True, basefmt=" ")
```

```python
sampledt, step_response = control.step_response(G,t)
```

```python
plt.stem(sampledt, step_response, use_line_collection=True, basefmt=" ")
```

```python
t = numpy.arange(0, 10+Ts, Ts)
u = numpy.sin(t)
sampledt, sine_response, input = control.forced_response(G,t,u)
```

```python
plt.step(sampledt, sine_response)
```
